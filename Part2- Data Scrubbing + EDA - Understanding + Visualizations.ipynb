Cleaning strings to include only numbers

```python
unpickled_start_page_df['Durations_spliced_str'] = unpickled_start_page_df['Durations_spliced'].apply(lambda x: x.split(' hour')[0])
unpickled_start_page_df['Durations_spliced_int'] = unpickled_start_page_df['Durations_spliced_str'].astype(float)
unpickled_start_page_df['Prices_spliced_str'] = unpickled_start_page_df['Prices_spliced_person'].apply(lambda x: x.split('$')[1])
unpickled_start_page_df['Prices_spliced_str'] = unpickled_start_page_df['Prices_spliced_str'].str.replace(',', '')
unpickled_start_page_df['Prices_spliced_int'] = unpickled_start_page_df['Prices_spliced_str'].astype(int)

unpickled_start_page_df['Ratings_int'] = unpickled_start_page_df['Ratings_spliced'].astype(float)
unpickled_start_page_df
```

Replace my strings 'n/a' with np.NaN & convert number objects to floats 

```python
import numpy as np
unpickled_start_page_replaced_nulls = unpickled_start_page_df.replace('n/a',np.NaN)
unpickled_start_page_replaced_nulls['Ratings_int'] = unpickled_start_page_replaced_nulls['Ratings_spliced'].astype(float)
unpickled_start_page_replaced_nulls.isna().sum()
```

Still have some nulls with the ratings (more than 2/3 of my observations!) which I decided to fill with the median. 

```python
unpickled_start_page_replaced_nulls['filled_na_Ratings_median'] = unpickled_start_page_replaced_nulls['Ratings_int'].fillna(unpickled_start_page_replaced_nulls['Ratings_int'].median())
```

```python
unpickled_start_page_replaced_nulls.info()
```

I was interested in the length of the Titles for each Experience--by length of the titles, average length of each word, and total number of characters. 

```python
unpickled_start_page_replaced_nulls['Title_word_length'] = unpickled_start_page_replaced_nulls['Titles'].str.split().str.len()
unpickled_start_page_replaced_nulls['Title_character_count'] = unpickled_start_page_replaced_nulls['Titles'].str.replace('.','').str.replace(' ','').str.len()
unpickled_start_page_replaced_nulls['Average_word_length'] = unpickled_start_page_replaced_nulls['Title_character_count']/unpickled_start_page_replaced_nulls['Title_word_length']
unpickled_start_page_replaced_nulls
```

Sadly, my original question of how these features impact Airbnb "guest" ratings was complicated by some limitations I observed in the ratings:
- Not only did > 2/3 of my observations have null ratings, but only 182 of the 558 listings had a rating below 5.0... so the ratings in my sample are mostly clustered at 5.0. This is great for the hosts, but not helpful for my model. I decided to go another route, but I did also consider possible reasons for this pattern and paths to resolve it.
