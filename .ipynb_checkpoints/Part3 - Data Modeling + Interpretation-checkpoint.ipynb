# MVP (Minimum Viable Product)

```python
import pandas as pd

import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

For my MVP, I first tried to model by rating--as I initially intended for my project. 

# Model by Rating

```python
unpickled_final_start_page_df = pd.read_pickle("./start_page.pkl")
unpickled_final_start_page_df.info()
X, y = unpickled_final_start_page_df.select_dtypes(exclude=['object']).drop('Ratings_int',axis=1).drop('Average_word_length',axis=1).drop('filled_na_Ratings_median',axis=1), unpickled_final_start_page_df['filled_na_Ratings_median']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=100)
print(X_train)
```

```python
lm_rating = LinearRegression()
lm_rating.fit(X_train,y_train)


print(f'Linear Regression val R^2 TRAIN (Rating): {lm_rating.score(X_train, y_train):.3f}')
```

```python
y_predict_rating = lm_rating.predict(X_test)
print(f'Linear Regression val R^2 test (Rating): {lm_rating.score(X_test, y_test):.3f}')
```

As seen in my EDA and confirmed in the graph below, most of the actual ratings are clustered at around 5.0. This narrow range of the target or dependent variable makes a linear regression analysis very difficult. Some possible ways to mitigate this could be: a Heckman correction--which can account for selection bias. Selection bias is very clearly affecting my results, which makes sense because ratings are essentially opt-in surveys for participants to submit voluntarily. People may not be as likely to write a review or give a rating if they had a less-than-stellar experience. There's also a large unrepresented population of people who may not have done the experience and high variability in the number of people who engage in each experience. Finally, there may be elements of survival bias in my data because businesses who do not do as well may stop offering the experience.  

```python
import matplotlib.pyplot as plt 
predict = y_predict_rating
actual = y_test
plt.scatter(predict, actual, alpha = 0.1)
plt.title('Scatter plot of y_predicted vs. y_actual for Airbnb Experience Ratings')
plt.xlabel('y_predicted')
plt.ylabel('y_actual');

```

Below I tested different polynomial feature transformations and combined them with standard scalers. I then tested different options with LassoCV and Ridge CV to find the best-fit model. SKLearn has several modules that can streamline this process! 

Note: Polynomial feature transformations are still considered LINEAR regressions because the coefficients remain linear. The units of the variables themselves add on additional units & terms as we add degrees (e.g., x1 becomes x1^2 + x1, etc.) The polynomial terms describe the non-linear relationship of these independent variables, but still use linear regression as the basis for the model. Polynomial transformations must occur BEFORE standard scaling and standard scaling MUST be done to do lasso and ridge regularizations. SKLearn has a great tool called lassoCV and ridgeCV--which perform cross validation across my designated number of k-folds, find the optimal alpha and coefficients, and produce my averge r_squared. My goal was to find a model that produces the highest r_squared, but not overfitting the data. 

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LassoCV
from sklearn.linear_model import RidgeCV
from sklearn.datasets import make_regression
from sklearn.metrics import mean_squared_error

#Polynomial degree 2 + StandardScaler + Ridge produced the best R^2 and minimizes mean square error, so I picked this model
lm_price_pipeline = make_pipeline(PolynomialFeatures(2), StandardScaler(), RidgeCV(cv=5,alphas=(0.01,0.1,1,10,100,1000,10000)))
#lm_price_pipeline = make_pipeline(PolynomialFeatures(4), StandardScaler(), LassoCV(cv=5,n_alphas=10))
lm_price_pipeline_trained = lm_price_pipeline.fit(X_train,y_train)
print(f'Pipeline val R^2 train: {lm_price_pipeline.score(X_train, y_train):.3f}')
print('Coefficients: '+str(list(zip(X_train.columns, lm_price_pipeline.steps[2][1].coef_))))
print(f'Pipeline alpha: {lm_price_pipeline.steps[2][1].alpha_:.3f}')
print('Alphas: '+str(lm_price_pipeline.steps[2][1].alphas))

y_predict_pipeline_train = lm_price_pipeline_trained.predict(X_train)
y_predict_pipeline_test = lm_price_pipeline_trained.predict(X_test)
print('MSE: '+str(mean_squared_error(y_train, y_predict_pipeline_train)))
#lm_price_pipeline.score(X_test,y_test)
```

```python
plt.scatter(y_predict_pipeline_test, y_test, alpha = 0.2)
plt.title('Predicted vs. Actual Airbnb Experience Prices')
plt.xlabel('Predicted Price')
plt.ylabel('Actual Price');
print(f'Pipeline val R^2 test: {lm_price_pipeline_trained.score(X_test, y_test):.3f}')
plt.plot(y_test, y_test, color='blue',linewidth=1);
plt.savefig("Y_predvY_test.png", format="png")
```

My model is underfitting--as indicated by the narrow range of y_predict values and the fact that my y_predict values are not lining up with the "optimal" regression line, where r^2 would = 1. This means my model's predictins are not correlating with the actual results of the prices.  

```python
residual = y_test - y_predict_pipeline_test
residual

plt.scatter(y_predict_pipeline_test,residual, alpha = 0.3)
plt.hlines (0, 3, 5)
plt.title('Scatter plot of y_predicted vs. residuals for Airbnb Experience Pricing')
plt.xlabel('y_predict_pipeline')
plt.ylabel('residuals');

#My residuals were relatively symmetrically distributed around zero (homoscedastic)
```
